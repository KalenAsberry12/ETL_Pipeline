```python
import pyodbc
from datetime import datetime, timedelta
import datetime
import csv
import pandas as pd
import cx_Oracle
import re
from config import connectionstring

def extract_transform():
    
    data = pd.read_csv(r'C:\Users\Kalen\DATA.csv')
    #rename column to merge later
    data = data.rename(columns={"email": "Email"})
    
    #dataframe to a list
    domainlist = data['Email'].values.tolist()
    
    #create domain dataframe to merge later 
    domaindf = pd.DataFrame()
    
    #run for loop to grab domains
    for x in domainlist:
        domain=re.findall('@+\S+[.in|.com|.uk]',x)[0]
        #print(domain)
        domaindf = domaindf.append({'Email': x, 'Domain': domain }, ignore_index=True)
    
    #merge data
    mergedData = pd.merge(data, domaindf, on ='Email')
    
    new_lst = []
    
    # looping through to append to new_list array
    # in order return to a variable 
    for index, row in mergedData.iterrows():
        newData = (row['id'], row['first_name'],row['last_name'], row['Email'],row['gender'], row['ip_address'],row['Domain'])
        #appending to new_lst to return to a variable 
        new_lst.append(newData)
    
    return new_lst

def insert_domain_data(new_lst):
    
    #query to insert into sql
    sql = ('insert into u6fz.cust_emails(id, first_name, last_name, Email, gender, ip_address, Domains)' 
          'values(:id, :first_name, :last_name, :Email, :gender, :ip_address, :Domain)')
    
    try:
        # establish a new connection
        with cx.connect(connectionstring) as connection:
            # create a cursor
            with connection.cursor() as cursor:
                
                #new_lst is variable
                cursor.executemany(sql, new_lst)
                
                #row = cursor.fetchone()
                
                connection.commit()
    except cx_Oracle.Error as error:
        print('Error occurred:')
        print(error)

def main():
    starttime=datetime.datetime.now()
    
    print("Program started at ...{}\n".format(starttime.strftime('%Y-%m-%d %H:%M:%S')))
    
    #data extract into pd, transformed and put into a lst
    new_lst = extract_transform()
    
    #load data function
    insert_domain_data(new_lst)
    
    #run time
    endtime=datetime.datetime.now()
    
    print("Run time: {}".format(endtime-starttime))

if __name__ == '__main__':
    main()
```
